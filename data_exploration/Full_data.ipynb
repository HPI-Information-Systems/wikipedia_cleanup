```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
pop_data = pd.read_csv('../data/popular_data.csv')
```

```python
pop_data
```

```python
pop_data.dropna(subset=['days_diff'],inplace=True)
```

```python
pop_data['days_diff'] = pop_data['days_diff'].apply(lambda x: pd.to_timedelta(x).days)
```

```python
top20_template_types_list = pop_data['template'].unique()
```

```python
top20_template_types_list
```

```python
for temp in top20_template_types_list:
    pop_data[pop_data['template']==temp].boxplot(column='days_diff', by=['property_name'], figsize=(15,7))
    plt.title('Boxplots for the template {}'.format(temp))
    plt.ylabel('Days until next change')
    plt.show()
```
